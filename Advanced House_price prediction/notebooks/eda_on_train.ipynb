```python
#Data analysis phase
#Main aim is to understand about data
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

#display all the columns of the dataframe
pd.pandas.set_option('display.max_columns', None)
```

```python
dataset = pd.read_csv('train.csv')

#print shape of dataset with rows and columns
print(dataset.shape)
```

```python
#print top 5 records
dataset.head()
```

###In data-analysis, we will analyze to find out the below stuff:

```python
#check the percentage of NaN values present in each feature
#1. make the list of features with missing values
features_with_na = [features for features in dataset.columns if dataset[features].isnull().sum() > 1]

#print the feature name and the percentage of missing values
for feature in features_with_na:
    print(feature, np.round(dataset[feature].isnull().mean(), 4), ' % missing values')

print(len(features_with_na))
```

Since there are many missing values, so we need to find relationship between missing values and salesprice
let's plot some diagrams for this relationship


```python
for feature in features_with_na:
    data = dataset.copy()

    #let's make a varible that indicates 1 if the observation was missing otherwise 0
    data[feature] = np.where(data[feature].isnull(), 1, 0)

    #let's calculate the mean salesprice where the information is missing or present
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
```

The relationship between the dependent variable (SalePrice) and missing values is clearly visible. So we need to replace these NaN values with something else, which will be done in feature engineering section.

```python
#List of numerical variables
numerical_features = [feature for feature in dataset.columns if dataset[feature].dtypes != 'O']
print("Number of numerical features", len(numerical_features))

#visualize the numerical variables
dataset[numerical_features].head()
```

Temporal Variables(Eg: Datetime Variables)
From the Dataset we have 4 year variables. We have extract information from the datetime variables like no of years or no of days. One example in this specific scenario can be difference in years between the year the house was built and the year the house was sold. We will be performing this analysis in the Feature Engineering which is the next video.

```python
#list of variables that contain year feature
year_feature = [feature for feature in numerical_features if 'Yr' in feature or 'Year' in feature ]
print(year_feature)
```

```python
#exploring content of year variables
for feature in year_feature:
    print(feature, dataset[feature].unique())
```

```python
#checking relationship between house sold year and Saleprice
dataset.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year Sold')
plt.ylabel('Sale price')
plt.title('House price vs Year Sold')
plt.show()
```

```python
#comparing difference between all year features and Sale Price
for feature in year_feature:
    if feature != 'YrSold':
        data = dataset.copy()
        #difference between year variable and the year house was sold 
        data[feature] = data['YrSold'] - data[feature]

        plt.scatter(data[feature], data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.show()
        
```

```python
discrete_feature = [feature for feature in numerical_features if len(dataset[feature].unique()) < 25 and feature not in year_feature +['Id']]
print(f"discrete variable count {len(discrete_feature)}")
```

```python
dataset[discrete_feature].head()
```

```python
#findind relationship between discrete features and SalePrice
for feature in discrete_feature:
    data =dataset.copy()
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title(feature)
    plt.show()
```

```python
continuous_feature = [feature for feature in numerical_features if feature not in discrete_feature + year_feature + ['Id']]
print(f"continuous feature count {len(continuous_feature)}")
```

```python
for feature in continuous_feature:
    data = dataset.copy()
    data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel('Saleprice')
    plt.title("SalePrice vs feature")
    plt.show()
```

```python
#since the continuous variables are right-tailed, we'll apply Log normal transformation
for feature in continuous_feature:
    data = dataset.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature] = np.log(data[feature])
        data['SalePrice'] = np.log(data['SalePrice'])
        plt.scatter(data[feature], data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.title(feature)
        plt.show()
```

```python
for feature in continuous_feature:
    data = dataset.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature] = np.log(data[feature])
        data.boxplot(column=feature)
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()
    
```
