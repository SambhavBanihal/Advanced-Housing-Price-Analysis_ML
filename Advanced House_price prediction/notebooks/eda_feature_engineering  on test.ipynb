```python
import numpy as np
import pandas as pd

def get_common_categories(train_df, categorical_features, threshold=0.01):
    """
    For each categorical feature, return a list of categories that occur
    in > threshold fraction of the train rows.
    """
    common_cats = {}
    n = len(train_df)
    for col in categorical_features:
        if col not in train_df.columns:
            common_cats[col] = []   # defensive; no such column in train
            continue
        freqs = train_df[col].value_counts(normalize=True)  # proportion
        common = freqs[freqs > threshold].index.tolist()
        common_cats[col] = common
    return common_cats

def replace_rare_categories(df, common_cats_map, rare_label='Rare_var'):
    """
    Replace categories not present in common_cats_map[col] with rare_label.
    Works on any dataframe (train or test), as long as columns exist.
    """
    df = df.copy()
    for col, common_list in common_cats_map.items():
        if col not in df.columns:
            # skip columns that don't exist in this dataframe
            continue
        if len(common_list) == 0:
            # if train had no frequent categories above threshold, keep original values
            # or optionally set everything to rare_label. Here we keep original.
            continue
        df[col] = np.where(df[col].isin(common_list), df[col], rare_label)
    return df
```

```python
# 1) prepare list of categorical features you want to treat
categorical_feature = [col for col in train.columns if train[col].dtype == "object"]

# 2) compute allowed (common) categories from train
common_cats_map = get_common_categories(train, categorical_feature, threshold=0.01)

# 3) apply the rule to train and test (so both have 'Rare_var' consistently)
train = replace_rare_categories(train, common_cats_map, rare_label='Rare_var')
test  = replace_rare_categories(test,  common_cats_map, rare_label='Rare_var')
```

```python
# --- Step 2: LOAD fitted objects from train and APPLY to test (corrected) ---
import joblib
import numpy as np
import pandas as pd

# file paths (adjust if you saved elsewhere)
SCALER_PATH = "scaler.pkl"
LE_DICT_PATH = "label_encoders.pkl"
NUM_FEAT_LIST = "numeric_features_list.pkl"
CAT_FEAT_LIST = "categorical_features_list.pkl"

# token you used in train for missing categorical values (must match train)
MISSING_TOKEN = "Missing"

# 1) load artifacts
scaler = joblib.load(SCALER_PATH)
le_dict = joblib.load(LE_DICT_PATH)          # dict: {col: LabelEncoder()}
numeric_features = joblib.load(NUM_FEAT_LIST)
categorical_features = joblib.load(CAT_FEAT_LIST)

# 2) intersect with columns present in test (avoid KeyErrors)
numeric_features_test = [c for c in numeric_features if c in test.columns]
categorical_features_test = [c for c in categorical_features if c in test.columns]

# 3) ensure test numeric columns are float to accept scaled values
test[numeric_features_test] = test[numeric_features_test].astype(float)

# 4) transform numeric features using scaler fitted on train
#    (scaler.transform returns floats; cast explicitly to float on assignment)
test.loc[:, numeric_features_test] = scaler.transform(test[numeric_features_test]).astype(float)

# 5) transform categorical features using saved LabelEncoders (no refit)
#    unseen categories in test -> mapped to -1
for col in categorical_features_test:
    le = le_dict.get(col)
    # defensive: if encoder missing, fill with -1
    if le is None:
        test[col] = test[col].fillna(MISSING_TOKEN).map({}).fillna(-1).astype(int)
        continue

    # replicate how you prepared train before encoding (fill NaNs with MISSING_TOKEN)
    test_col_filled = test[col].fillna(MISSING_TOKEN).astype(object)

    # build mapping from encoder classes_ -> integer labels used by the encoder
    mapping = {cat: code for code, cat in enumerate(le.classes_)}

    # map test values; unseen -> NaN -> fill with -1
    test[col] = test_col_filled.map(mapping).fillna(-1).astype(int)

# 6) quick sanity checks (optional)
print("Numeric cols transformed (count):", len(numeric_features_test))
print("Categorical cols encoded (count):", len(categorical_features_test))
print("Any remaining NaNs in transformed numeric cols?", test[numeric_features_test].isnull().sum().sum())
print("Example - unseen category counts per categorical (should be >=0):")
for c in categorical_features_test[:10]:
    print(c, (test[c] == -1).sum())
```
